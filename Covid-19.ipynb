```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy
import pandas as pd

sns.set(font_scale = 1.2)
```

```python
# Starting Feb 24
new_ontario_march = np.array([4,3,2,0,3,5,0,4,3,1,6,17,20,24,42,32,12,25,43])
new_canada_march = np.array([4,3,3,4,13,7,6,7,11,18,21,40,39,54,74,101,144,121,81])

cases_canada_march = np.cumsum(new_canada_march)
cases_ontario_march = np.cumsum(new_ontario_march)
```

```python
def exponential_growth(t,r0,r):
    return r0 * (1 + r) ** t
```

# Ontario Predicted Cases (March)

```python
x_ontario_march = np.linspace(1,len(cases_ontario_march),len(cases_ontario_march))

params = scipy.optimize.curve_fit(exponential_growth, x_ontario_march, cases_ontario_march, p0=(4,0.5))
r0,r = params[0]

n = len(cases_canada_march) + 1

n_pred = n + 1
x_pred = np.linspace(1, n_pred, n_pred)
y_pred = exponential_growth(x_pred,r0,r)

print(f"r0:{r0} r:{r}")
print(f"Current Total: {cases_ontario_march[-1]}")
print(f"New Cases Ontario Tomorrow: {y_pred[-1] - cases_ontario_march[-1]}")
plt.bar(x_ontario_march, cases_ontario_march)
plt.plot(x_pred, y_pred, c="r")
```

# Canada Predicted Cases (March)

```python
x_canada_march = np.linspace(1,len(cases_canada_march), len(cases_canada_march))

params = scipy.optimize.curve_fit(exponential_growth, x_canada_march, cases_canada_march, p0=(4,0.5))
r0,r = params[0]

n = len(cases_canada_march) + 1

n_pred = n + 1
x_pred = np.linspace(1, n_pred, n_pred)
y_pred = exponential_growth(x_pred,r0,r)

print(f"r0:{r0} r:{r}")
print(f"Current Total: {cases_canada_march[-1]}")
print(f"New Cases Canada Tomorrow: {y_pred[-1] - cases_canada_march[-1]}")
plt.bar(x_canada_march, cases_canada_march)
plt.plot(x_pred, y_pred, c="r")
```
